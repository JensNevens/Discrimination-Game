```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data.csv')
plt.figure()
plt.title('The Discrimination Game')

ax1 = df.S.plot(label='Rep. Size')
ax1.set_ylabel('Repertoire Size')
ax1.legend(loc=4)

ax2 = df.R.plot(secondary_y=True, style='g', label='Success')
ax2.set_ylabel('Success Rate')
ax2.legend(loc=3)

#plt.savefig('./plots/Age150_Success0075_Freq25.pdf')
```
